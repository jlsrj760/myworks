#Finding Placement Status

```python
#Impoerting Library
import pandas as pd
from sklearn.model_selection import train_test_split
#ML algorithh KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
#ML Algorithm
from sklearn.ensemble import AdaBoostClassifier
```

```python
#importing Data Set
df = pd.read_csv('/content/01 Train Data.csv')
```

```python
#converting traing data sets status into Binary Values
df['Placement Status 0 or 1'] = df['Placement Status'].map({'Placed': int(0), 'Not placed': int(1)})
df['Placement Status 0 or 1'] = df['Placement Status 0 or 1'].apply(lambda x: 1 if x > 0 else 0)
df.tail()
```

```python
#Selecting Features
selected_features = ['CGPA', 'Speaking Skills', 'ML Knowledge']
```

```python
X = df[selected_features]
y = df['Placement Status 0 or 1']
```

```python
#Creating data set for training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
model  =    KNeighborsClassifier(n_neighbors=3)
```

```python
model =  AdaBoostClassifier(n_estimators=50, random_state=42)
```

```python
#Training and Testing model
model.fit(X_train,y_train)
predictions = model.predict(X_test)
```

```python
#checking Accuracy of the modl
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy:.2f}')
```
